# 3. パラメトリックコピュラ

### **図3.1**   FGMコピュラの密度

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")

# FGMコピュラの密度関数
def FGM_copula_pdf(u, v, theta=1.0):
  return 1 + theta * (1 - 2*u) * (1 - 2*v)

# データの生成
u = np.linspace(0.05, 0.95, 50)
v = np.linspace(0.05, 0.95, 50)
U, V = np.meshgrid(u, v)

# グラフの作成
fig_contour = plt.figure(figsize=(12, 6))

ax1 = fig_contour.add_subplot(1, 2, 1)
contour_lower = ax1.contour(U, V, FGM_copula_pdf(U, V, -1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_lower, ax=ax1)
ax1.set_xlabel('u')
ax1.set_ylabel('v')
ax1.set_title('FGM Copula (theta=-1.0)')
ax1.set_aspect('equal', adjustable='box')

ax2 = fig_contour.add_subplot(1, 2, 2)
contour_upper = ax2.contour(U, V, FGM_copula_pdf(U, V, 1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_upper, ax=ax2)
ax2.set_xlabel('u')
ax2.set_ylabel('v')
ax2.set_title('FGM Copula (theta=1.0)')
ax2.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

### **図3.2**   AMHコピュラの密度

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")

# AMHコピュラの密度関数
def AMH_copula_pdf(u, v, theta=0.5):
  return (1 + theta*((1+u)*(1+v)-3) + theta**2 * (1-u)*(1-v)) / (1 - theta * (1-u) * (1-v))**3

# データの生成
u = np.linspace(0.05, 0.95, 100)
v = np.linspace(0.05, 0.95, 100)
U, V = np.meshgrid(u, v)

# グラフの作成
fig_contour = plt.figure(figsize=(12, 6))

ax1 = fig_contour.add_subplot(1, 2, 1)
contour_lower = ax1.contour(U, V, AMH_copula_pdf(U, V, -1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_lower, ax=ax1)
ax1.set_xlabel('u')
ax1.set_ylabel('v')
ax1.set_title('AMH Copula (theta=-1.0)')
ax1.set_aspect('equal', adjustable='box')

ax2 = fig_contour.add_subplot(1, 2, 2)
contour_upper = ax2.contour(U, V, AMH_copula_pdf(U, V, 1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_upper, ax=ax2)
ax2.set_xlabel('u')
ax2.set_ylabel('v')
ax2.set_title('AMH Copula (theta=1.0)')
ax2.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

### **図3.3**   GBコピュラの密度

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")

# GBコピュラの密度関数
def GB_copula_pdf(u, v, theta=0.5):
  return (1- theta - theta * np.log(u*v) + theta**2 * np.log(u) * np.log(v)) * np.exp(- theta * np.log(u) * np.log(v))

# データの生成
u = np.linspace(0.05, 0.95, 100)
v = np.linspace(0.05, 0.95, 100)
U, V = np.meshgrid(u, v)

# グラフの作成
fig_contour = plt.figure(figsize=(12, 6))

ax1 = fig_contour.add_subplot(1, 2, 1)
contour_lower = ax1.contour(U, V, GB_copula_pdf(U, V, 0.5), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_lower, ax=ax1)
ax1.set_xlabel('u')
ax1.set_ylabel('v')
ax1.set_title('GB Copula (theta=0.5)')
ax1.set_aspect('equal', adjustable='box')

ax2 = fig_contour.add_subplot(1, 2, 2)
contour_upper = ax2.contour(U, V, GB_copula_pdf(U, V, 1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_upper, ax=ax2)
ax2.set_xlabel('u')
ax2.set_ylabel('v')
ax2.set_title('GB Copula (theta=1.0)')
ax2.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

### **図3.4**   CCコピュラの密度

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")

# CCコピュラの密度関数
def CC_copula_pdf(u, v, theta=0.5):
  return (1+ (3*u*v - u - v)* theta + u*(1-u) * v*(1-v) * theta**2 ) * np.exp(theta * (1-u) * (1-v))

# データの生成
u = np.linspace(0.05, 0.95, 100)
v = np.linspace(0.05, 0.95, 100)
U, V = np.meshgrid(u, v)

# グラフの作成
fig_contour = plt.figure(figsize=(12, 6))

ax1 = fig_contour.add_subplot(1, 2, 1)
contour_lower = ax1.contour(U, V, CC_copula_pdf(U, V, -1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_lower, ax=ax1)
ax1.set_xlabel('u')
ax1.set_ylabel('v')
ax1.set_title('CC Copula (theta=-1.0)')
ax1.set_aspect('equal', adjustable='box')

ax2 = fig_contour.add_subplot(1, 2, 2)
contour_upper = ax2.contour(U, V, CC_copula_pdf(U, V, 1.0), levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_upper, ax=ax2)
ax2.set_xlabel('u')
ax2.set_ylabel('v')
ax2.set_title('CC Copula (theta=1.0)')
ax2.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```
