# 2. コピュラの定義と基本定理

### **図2.2**  独立コピュラのグラフ

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
sns.set_style("darkgrid")

# 独立コピュラ関数
def independent_copula(u, v):
  return u * v

# データの生成
u = np.linspace(0, 1, 50)
v = np.linspace(0, 1, 50)
U, V = np.meshgrid(u, v)
Z = independent_copula(U, V)

# グラフの描画
fig = plt.figure(figsize=(12, 5))

# 3次元グラフ
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(V, U, Z, cmap='PuBu_r',edgecolor='black')
ax1.set_xlabel('u')
ax1.set_ylabel('v')
ax1.set_zlabel('C(u, v)')
ax1.set_title('Independent Copula (3D)')

# 等高線グラフ
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contour(U, V, Z, levels=20, cmap='PuBu_r')
fig.colorbar(contour, ax=ax2)
ax2.set_xlabel('u')
ax2.set_ylabel('v')
ax2.set_title('Independent Copula (Contour)')
ax2.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

### **図2.4** 下界と上界コピュラの3次元グラフ

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")

# 下界コピュラ関数
def lower_bound_copula(u, v):
  return np.maximum(u + v - 1, 0)

# 上界コピュラ関数
def upper_bound_copula(u, v):
  return np.minimum(u, v)

# データの生成
u = np.linspace(0, 1, 50)
v = np.linspace(0, 1, 50)
U, V = np.meshgrid(u, v)

Z_lower = lower_bound_copula(U, V)
Z_upper = upper_bound_copula(U, V)

# グラフの描画
fig = plt.figure(figsize=(12, 6))

# 下界コピュラの3次元グラフ
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(U, V, Z_lower, cmap='PuBu_r',edgecolor='black')
ax1.set_xlabel('u')
ax1.set_ylabel('v')
ax1.set_zlabel('C(u, v)')
ax1.set_title('Lower Bound Copula (3D)')
ax1.set_zlim(0, 1)  # z軸の範囲を0から1に設定

# 上界コピュラの3次元グラフ
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(U, V, Z_upper, cmap='PuBu_r',edgecolor='black')
ax2.set_xlabel('u')
ax2.set_ylabel('v')
ax2.set_zlabel('C(u, v)')
ax2.set_title('Upper Bound Copula (3D)')
ax2.set_zlim(0, 1)  # z軸の範囲を0から1に設定


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

# 下界コピュラの等高線グラフ
fig_contour = plt.figure(figsize=(12, 6))

ax3 = fig_contour.add_subplot(1, 2, 1)
contour_lower = ax3.contourf(U, V, Z_lower, levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_lower, ax=ax3)
ax3.set_xlabel('u')
ax3.set_ylabel('v')
ax3.set_title('Lower Bound Copula (Contour)')
ax3.set_aspect('equal', adjustable='box')

# 上界コピュラの等高線グラフ
ax4 = fig_contour.add_subplot(1, 2, 2)
contour_upper = ax4.contourf(U, V, Z_upper, levels=20, cmap='PuBu_r')
fig_contour.colorbar(contour_upper, ax=ax4)
ax4.set_xlabel('u')
ax4.set_ylabel('v')
ax4.set_title('Upper Bound Copula (Contour)')
ax4.set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```
